# 배경 및 목적

현대 민주주의는 대의 민주주의(representative democracy)이며, 그 중심에는 유권자에 의해 선출되어 위임받은 권한을 행사하는 대표자(representatives)가 존재합니다. 한정된 자원과 시간 등의 문제로 직접 민주주의가 불가능한 지금, 대의 민주주의는 현실적으로 선택가능한 제도라고 할 수 있습니다. 그렇다면 우리의 고민은 우리사회의 대의 민주주의가 실현되는 공간, 국회가 과연 유권자들의 기대에 부응하고 있는지, 혹은 과거에는 부족했을지라도 점차 나아지고 있는지, 그러한 변화가 실질적인 것인지 표면적인 것에 불과하는지를 이해하는 것입니다.

국회의 입법활동은 이러한 대의 민주주의의 작동원리와 성과를 직접적으로 보여주는 정치적 행위입니다. 그럼에도 불구하고 국회의 입법활동은 전문적인 용어들과 법률 조항, 제도들로 인하여 일반 시민들이 쉽게 접근하기 어려운 영역에 머물러 있습니다. 우리 국회는 국회 내 회의들의 속기록 및 의사결정 사항들을 다양한 방식으로 중계하고 공개하여 그 투명성을 날로 제고하고 있지만, 정작 입법 과정을 이해하고 실생활과 연관 지어 생각하는 것은 쉽지 않습니다.

# 1. 각 대수 국회 입활활동 평가



먼저 국회의원들이 본연의 역할이라고 할 수 있는 **입법활동을 얼마나 많이 했는지**를 살펴보고자 합니다. 여기서는 각 대수별로 국회의원들이 법안을 발의한 수를 살펴보고자 합니다. 국회의 입안 수는 일반적으로 입법생산성을 보여주는 지표로 평가됩니다.

### 각 대수별 국회의원 수

우리나라 국회 대수별 국회의원 수를 짚어보고 넘어가겠습니다.

```r
# options(repr.plot.width = 3, repr.plot.height = 2, repr.plot.res = 300)
options(repr.plot.width=8, repr.plot.height=5)
people_data %>%
    select(DAESU:BIRTH) %>%
    filter(DAESU >= 15) %>%
    mutate(DAESU = paste(DAESU, "대")) %>%
    group_by(DAESU) %>%
    summarize(n = n()) %>%
    ggplot(., aes(x = as.factor(DAESU), n)) +
    geom_bar(stat = "identity", col = "gray10", width = 0.75, fill = "orangered3") +
    geom_text(aes(y = n + 15, label = n), size = 12, family="notosanskr")+
    ylim(NA, 400) + 
    theme(legend.position = "none",
          panel.grid.major.x=element_blank(),
          title = element_text(size = 50, family="notosanskr"),
          axis.title = element_text(size = 30, family="notosanskr"), 
          axis.text = element_text(size = 30, family="notosanskr"), 
          axis.line = element_line(size = 0.4, colour = "grey10")) +
    labs(title = "국회 대수별 국회의원 수",
         x = "", y = "인원 수 (명)")
```

위의 그래프는 각 대수 별 국회의원 수의 변화를 보여줍니다. 국회의원의 정수는 헌법에 명시되어 있지만, **선거법 위반 혹은 국회의원 개인의 일신상의 이유로 의원의 역할을 수행할 수 없을 때, 해당 의원의 자리를 재선거 혹은 보궐선거의 방식으로 교체하게 됩니다.** 따라서 대수별로 국회의원 정수에는 차이가 존재할 수 있습니다. 그래프에서 우리는 제15대에 국회의원의 수가 가장 많았으며, 현재 개원한 21대를 제외하고는 **제16대 국회의 의원 수가 정수에서 가장 변화가 적다**는 것을 보여줍니다.

### 각 대수 별로 국회는 얼마나 일을 많이 했을까

그렇다면 각 대수 별로 국회는 얼마나 일을 많이 했을까요? 다음의 그래프는 제15대부터 제20대 사이의 국회의원들의 법안발의 수를 대수별로 보여줍니다.

```r
suggest_data %>% 
    group_by(AGE) %>% 
    dplyr::filter(AGE >= 15) %>%
    summarise(n = n()) %>% 
    ggplot(., aes(x = factor(AGE, labels = paste0(15:21, "대")),
                  y = n)) +
    geom_bar(stat = "identity", col = "gray10", width = 0.75, fill = "orangered3") +
    geom_text(aes(y = n + 1000, label = n), size = 12, family = "notosanskr") +
    theme(legend.position = "none",
          panel.grid.major.x=element_blank(),
          title = element_text(size = 50, family="notosanskr"),
          axis.title = element_text(size = 30, family="notosanskr"), 
          axis.text = element_text(size = 30, family="notosanskr"), 
          axis.line = element_line(size = 0.4, colour = "grey10")) +
    labs(title = "대수별 국회 법안 발의건수 (15대~21대)",
        x = "", y = "발의건수")
```

위의 그래프에서 우리는 대수를 거듭할수록 국회의원들의 법안 발의 수의 총량이 큰 폭으로 증가하는 것을 확인할 수 있습니다. 

* 제20대 국회는 총 2,1594 건의 가장 높은 입법 발의건수를 보여주고 있습니다.
* 제21대 국회의 경우 2020년 6월 5일에 개원하여 약 6개월 동안 4,015 건을 발의한 것을 확인할 수 있습니다.

제21대 국회의 폐원일이 2024년 5월 29일이라는 점을 감안해 볼 때, 제21대 국회는 제20대 국회와 비슷하거나 더 많은 발의가 이루어질 수 있다고 가늠해 볼 수 있습니다.

### 대수별 국회의원 1인 당 법안 발의건수

앞에서 살펴보았듯 국회의원의 수는 국회 대수별로 다르기 때문에 우리는 국회의원 1명을 기준으로 보았을 때, 각 국회별 국회의원 인당 법안 발의수를 살펴보았습니다. 

```r
member <- people_data %>%
    filter(DAESU >= 15) %>%
    group_by(DAESU) %>%
    summarize(n = n()) %>% 
    select(DAESU, MEM_NM = n)

sum_suggest <- suggest_data %>% 
    group_by(AGE) %>% 
    dplyr::filter(AGE >= 15) %>%
    summarise(n = n()) 

sum_member <- left_join(sum_suggest, member, by = c("AGE" = "DAESU"))
    
sum_member %>%
    mutate(LawPer = round(n / MEM_NM, 1)) %>%
    ggplot(., aes(x = factor(AGE, labels = paste0(15:21, "대")),
                  y = LawPer)) +
    geom_bar(stat = "identity", col = "gray10", width = 0.75, fill = "orangered3") +
    geom_text(aes(y = LawPer + 3, label = LawPer), size = 12, family = "notosanskr") +
    theme(legend.position = "none",
          panel.grid.major.x=element_blank(),
          title = element_text(size = 50, family="notosanskr"),
          axis.title = element_text(size = 30, family="notosanskr"), 
          axis.text = element_text(size = 30, family="notosanskr"), 
          axis.line = element_line(size = 0.4, colour = "grey10")) +
    labs(title = "대수별 국회의원 1인 당 법안 발의건수 (15대~21대)",
        x = "", y = "발의건수")
```

위의 그래프에서 우리는 다음과 같은 사실을 확인 할 수 있었습니다.

* 제20대 국회에 이르러서는 국회의원 1인당 법안발의 건수가 제15대에 비해 약 27배 증가한 것을 확인할 수 있습니다. 
* 제20대 국회의 국회의원 1인당 법안발의 건수는 직전의 제19대와 비교하여서도 약 1.5배에 달합니다.

이제 막 개원한 제21대 국회를 제외한다면, 국회의원들이 시간이 갈수록 더 왕성한 입법활동을 수행하고 있다는 것을 확인할 수 있습니다.

# 전체 발의안

전체 발의된 안건 62,658건 중 이제까지 15대 국회에서부터 발의 된 안건들, 60,428건을 가지고 각 국회의 활동성 평가를 해보도록 하겠습니다.

```r
# 15대 국회 1996년 5월 30일 개원
suggest_data <- suggest_data %>%
    mutate(Year = format(as.Date(PROPOSE_DT), "%Y") %>% as.integer(),
           Month = format(as.Date(PROPOSE_DT), "%m") %>% as.integer()/12,
           datenum = Year + Month) %>% 
    filter(datenum > 1996.5)
dim(suggest_data)[1]
```

# 국회의원들은 1년 중 언제 일을 가장 열심히 할까?

상대적으로 가장 시각화가 쉬웠던 월별 입법 발의 건수에 대하여 살펴보도록 하겠습니다. 1월에서 5월 사이에 발의 건수 대비 국회가 개원하는 6월부터 연말까지의 발의 건수가 훨씬 많습니다. 아무래도 선거가 있는 3월과 4월에 밀렸던 안건들, 그리고 새로운 국회가 시작되어 열심히 하고자 하는 의지가 반영된 것으로 보이네요. 기간은 1년을 기준으로 살펴보는 것이므로 1996년 6월부터 2020년 5월 사이의 자료를 기준으로 시각화 하였습니다. 

```r
suggest_data %>% 
    filter(datenum <= 2020.5) %>%
    mutate(Month = format(as.Date(PROPOSE_DT), "%m")) %>%
    group_by(Month) %>% 
    summarise(n = n()) %>% 
    ggplot(., aes(x = factor(Month, labels = paste0(1:12, "월")),
                  y = n)) +
    geom_bar(stat = "identity", col = "gray10", width = 0.75, fill = "orangered3") +
    geom_text(aes(y = n + 200, label = n), size = 12, family = "notosanskr") +
    theme(legend.position = "none",
          panel.grid.major.x=element_blank(),
          title = element_text(size = 50, family="notosanskr"),
          axis.title = element_text(size = 30, family="notosanskr"), 
          axis.text = element_text(size = 30, family="notosanskr"), 
          axis.line = element_line(size = 0.4, colour = "grey10")) +
    labs(title = "월별 국회 법안 발의건수 (15대~20대)",
        x = "", y = "발의건수")
```
